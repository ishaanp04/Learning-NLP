```python
# Wordnet Lemmatizer
# After lemmatization, the output we get is a valid word that means the same thing as the input. The output is called a root word
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize('going', pos='v'), lemmatizer.lemmatize('going'), lemmatizer.lemmatize('history'), lemmatizer.lemmatize('goes'), lemmatizer.lemmatize('congratulation')
```

```python
words = ['eating', 'eats', 'eaten', 'writing', 'writes', 'written', 'wrote', 'programming', 'programs', 'programmer', 'history']
```

```python
for word in words:
  print(word + ': ' + lemmatizer.lemmatize(word, pos='v '))
```

```python
lemmatizer.lemmatize("fairly", pos='n'), lemmatizer.lemmatize("sportingly", pos='n')
```
